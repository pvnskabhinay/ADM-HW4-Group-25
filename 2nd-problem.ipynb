```python
import json
from pprint import pprint
data = json.load(open('reduced_dblp.json'))
```

```python
publications = {}
for i in range(len(data)):
    lst = []
    for j in data[i]['authors']:
        lst.append(j['author_id'])
    publications[data[i]['id_publication_int']] = lst
```

```python
from collections import defaultdict
authors = defaultdict(list)

for key, values in publications.items():
    for value in values:
        authors[value].append(key)
```

```python
def jaccard_distance(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(list1) + len(list2)) - intersection
    if int(union)==0:
        return 0
    return 1-float(intersection / union)
```

```python
import itertools
import networkx as nx
graph = nx.Graph()

for pub, aut in publications.items():
    if len(aut)>1:
        combos = list(itertools.combinations(aut,2))
        for names in combos:
            graph.add_edge(names[0], names[1], weight = jaccard_distance(authors[names[0]], authors[names[1]]))
    else:
        graph.add_node(aut[0])
```

```python
import itertools
import networkx as nx
graph_names = nx.Graph()

for pub, aut in publications_.items():
    if len(aut)>1:
        combos = list(itertools.combinations(aut,2))
        for names in combos:
            graph_names.add_edge(names[0], names[1], weight = 1 )
    else:
        graph_names.add_node(aut[0])
```

```python
###Function for get the list of author that have at most the distance hop d
def getauthors(graph, author, d, aux, target_authors = []):
    graph=nx.Graph(graph)
    for each in graph.neighbors(author):
        aux+=1
        if aux<=int(d):
            if each not in target_authors: 
                target_authors.append(each)
                getauthors(graph, each, d, aux, target_authors)
            else:
                continue
            return target_authors
```
